```python
def calculate_rmse(actual, predicted):
    return np.sqrt(np.mean((actual - predicted) ** 2))
rmse_rf = calculate_rmse(hybrid_prediction['Rent'], hybrid_prediction['RandomForest_pred'])
rmse_lgbm = calculate_rmse(hybrid_prediction['Rent'], hybrid_prediction['LGBM_PRED'])
rmse_hybrid = calculate_rmse(hybrid_prediction['Rent'], hybrid_prediction['HYBRID_PREDICTION'])
rmse_df = pd.DataFrame({
    'Model': ['Random Forest', 'LightGBM', 'Hybrid'],
    'RMSE': [rmse_rf, rmse_lgbm, rmse_hybrid]
})
print(rmse_df)
```

```python
from sklearn.metrics import r2_score
r2_rf = r2_score(hybrid_prediction['Rent'], hybrid_prediction['RandomForest_pred'])
r2_lgbm = r2_score(hybrid_prediction['Rent'], hybrid_prediction['LGBM_PRED'])
r2_hybrid = r2_score(hybrid_prediction['Rent'], hybrid_prediction['HYBRID_PREDICTION'])
r2_df = pd.DataFrame({
    'Model': ['Random Forest', 'LightGBM', 'Hybrid'],
    'R² Score': [r2_rf, r2_lgbm, r2_hybrid]
})
print(r2_df)
```
